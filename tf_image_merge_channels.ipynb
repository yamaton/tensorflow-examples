## Merge images with TensorFlow

Will show how two images are merged channel-wise in TensorFlow.

```python
import imageio
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = imageio.imread('https://raw.githubusercontent.com/imageio/imageio-binaries/master/images/astronaut.png')
plt.imshow(img1[:,:,0], cmap='gray')
```

```python
img2 = imageio.imread('https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png')
plt.imshow(img2[:,:,0], cmap='gray')
```

```python
img3 = imageio.imread('https://upload.wikimedia.org/wikipedia/commons/2/25/Blisk-logo-512-512-background-transparent.png')
plt.imshow(img3[:,:,0], cmap='gray')
```

## Merge images channel-wise using TensorFlow

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
img1_tensor = tf.constant(img1[:,:,0])
img2_tensor = tf.constant(img2[:,:,0])
img3_tensor = tf.constant(img3[:,:,0])
```

```python
merged = tf.stack([img1_tensor, img2_tensor, img3_tensor], axis=-1)
merged.shape
```

```python
plt.imshow(merged)
```

## Merge images channel-wise using numpy

```python
import numpy as np
```

```python
merged_np = np.stack([img1[:,:,0], img2[:,:,0], img3[:,:,0]], axis=-1)
merged_np.shape
```

```python
plt.imshow(merged_np)
```
